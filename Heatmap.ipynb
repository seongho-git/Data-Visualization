```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    '1': [99.91, 99.63, 98.51, 93.97, 75.58],
    '4': [99.90, 99.61, 98.43, 93.71, 74.90],
    '16': [99.90, 99.60, 98.36, 93.66, 74.81],
    '64': [99.89, 99.57, 98.29, 93.18, 72.30],
    '256': [99.87, 99.47, 97.97, 91.79, 69.07]
}

df = pd.DataFrame(data, index=[1024, 256, 64, 16, 4])

# Draw a heatmap with the numeric values in each cell
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, fmt=".2f", cmap='Blues', linewidths=.5)
plt.title('Summarization and Generation Stage Ratio \n\n', fontsize=12)
plt.text(0.5, 1.03, 'Model : Gemma-2B, Batch : 1', ha='center', fontsize=10, transform=plt.gca().transAxes)
plt.xlabel('Input Length')
plt.ylabel('Output Length')
plt.show()
```
